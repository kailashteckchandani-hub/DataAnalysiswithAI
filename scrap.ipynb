So this code is converting messy nested API JSON into a clean structure that can be analyzed, stored, or exported to CSV.



```python
import requests
import pandas as pd

url = "https://www.nseindia.com/api/NextApi/apiClient/marketWatchApi?functionName=getIndicesData&symbol=NIFTY%2050"

session = requests.Session()
headers = {
    "accept": "application/json, text/plain, */*",
    "user-agent": "Mozilla/5.0",
    "referer": "https://www.nseindia.com/"
}

response = session.get(url, headers=headers, timeout=30)
json_data = response.json()

detail_df = pd.json_normalize(json_data["data"]["data"])
summary_df = pd.json_normalize({
    "timestamp": json_data["data"]["timestamp"],
    **json_data["data"]["aduCount"],
    **json_data["data"]["marketStatus"]
})

detail_df.to_csv("nse_market_data.csv", index=False)
summary_df.to_csv("nse_summary.csv", index=False)

print(detail_df.head())
print(summary_df.head())
```

```python
import os
import requests
import pandas as pd
from datetime import datetime

DATA_DIR = "output"
DETAIL_FILE = os.path.join(DATA_DIR, "nse_market_data_history.csv")
SUMMARY_FILE = os.path.join(DATA_DIR, "nse_summary_history.csv")

os.makedirs(DATA_DIR, exist_ok=True)

url = "https://www.nseindia.com/api/NextApi/apiClient/marketWatchApi?functionName=getIndicesData&symbol=NIFTY%2050"

headers = {
    "accept": "application/json, text/plain, */*",
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36",
    "referer": "https://www.nseindia.com/",
}

session = requests.Session()
response = session.get(url, headers=headers, timeout=30)
response.raise_for_status()

json_data = response.json()

detail_df = pd.json_normalize(json_data["data"]["data"])
detail_df["snapshot_date"] = datetime.now().strftime("%Y-%m-%d")
detail_df["snapshot_time"] = datetime.now().strftime("%H:%M:%S")

summary_df = pd.json_normalize({
    "timestamp": json_data["data"].get("timestamp"),
    **json_data["data"].get("aduCount", {}),
    **json_data["data"].get("marketStatus", {})
})
summary_df["snapshot_date"] = datetime.now().strftime("%Y-%m-%d")
summary_df["snapshot_time"] = datetime.now().strftime("%H:%M:%S")

if os.path.isfile(DETAIL_FILE):
    old_detail = pd.read_csv(DETAIL_FILE)
    combined_detail = pd.concat([old_detail, detail_df], ignore_index=True)
    if "snapshot_date" in combined_detail.columns and "symbol" in combined_detail.columns:
        combined_detail = combined_detail.drop_duplicates(subset=["snapshot_date", "symbol"], keep="last")
    combined_detail.to_csv(DETAIL_FILE, index=False)
else:
    detail_df.to_csv(DETAIL_FILE, index=False)

if os.path.isfile(SUMMARY_FILE):
    old_summary = pd.read_csv(SUMMARY_FILE)
    combined_summary = pd.concat([old_summary, summary_df], ignore_index=True)
    if "snapshot_date" in combined_summary.columns:
        combined_summary = combined_summary.drop_duplicates(subset=["snapshot_date"], keep="last")
    combined_summary.to_csv(SUMMARY_FILE, index=False)
else:
    summary_df.to_csv(SUMMARY_FILE, index=False)

print("Saved:", DETAIL_FILE)
print("Saved:", SUMMARY_FILE)
print("Detail rows:", len(detail_df))
print("Summary rows:", len(summary_df))
```
